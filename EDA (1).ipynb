```python
#data prepration 

df.drop(['Opening date'], axis=1)
```

```python
df = df[['coaster_name', 
         'Location', 
         'Status', 
         'Manufacturer',
         'year_introduced',
         'latitude', 
         'longitude',
         'Type_Main',
         'opening_date_clean',
         'speed_mph', 
         'height_ft',
         'Inversions_clean', 
         'Gforce_clean']].copy()
```

```python
df['opening_date_clean'] = pd.to_datetime(df['opening_date_clean'])

#functon can converts a scalar arrary like or series or data frame dictionary like to a pandas
```

```python
df = df.rename(columns={'coaster_name' : 'Coaster_Name',
                        'year_introduced' : 'Year_Introduced',
                        'opening_date_clean' : 'Opening_Date',
                        'speed_mph' : 'Speed_mph',
                        'height_ft' : 'Height_ft',
                        'Inversions_clean' : 'Inversions',
                        'Gforce_clean' : 'Gforce'})
```

```python
df = df.loc[ ~df.duplicated(subset=['Coaster_Name', 'Location', 'Opening_Date'])].reset_index(drop=True).copy()

#remove duplicates based on selected columns
```

```python
df['Year_Introduced'].value_counts()
```

```python
ax = df["Year_Introduced"].value_counts().head(10).plot(kind='bar' , title='Top 10 Year Coaster')
ax.set_xlabel('Year Intorduced')
ax.set_ylabel('Count')
```

```python
ax = df['Speed_mph'].plot(kind='hist',
                          bins=20,
                          title = 'Coaster Speed')
ax.set_xlabel('speed mph')
```

```python
ax = df["Speed_mph"].plot(kind='kde',
                          title = 'Coster speed')
ax.set_xlabel('Speed')
#Kernel Density Estimation
```

```python
df.plot(kind='scatter',
        x='Speed_mph',
        y='Height_ft',
        title = 'Cosaster vs height')
plt.show()
```

```python
ax = sns.scatterplot(x='Speed_mph',
                     y='Height_ft',
                     hue="Year_Introduced",
                     data = df)
ax.set_title('coaster speed vs height')
plt.show()
```

```python
sns.pairplot(df,
             vars=['Year_Introduced', 'Speed_mph', 'Height_ft', 'Inversions', 'Gforce'],
             hue = 'Type_Main')
plt.show()
```

```python
df_corr = df[['Year_Introduced', 'Speed_mph', 'Height_ft', 'Inversions', 'Gforce']].dropna().corr()
df_corr
#It measures the strength and direction of relationship between two numerical variables.
```

```python
sns.heatmap(df_corr, annot=True)
```
