```python
import numpy as np
#for numerical computing for arrya operation and mathematical function liner algebra
import pandas as pd 
#using for data manipulation and analysis easy to use data structure and analysis 
from sklearn.model_selection import train_test_split
#Provides a wide range of tools and algorithms for various Ml task includeing classification regression and clustering
#train test slipt function help in splitting the data set into training and the testing subset
from sklearn.feature_extraction.text import TfidfVectorizer
#TFID Vector feature of extraction commonly usedd in NLP and text mining task convert of text doc into a numerical representation 
from sklearn.linear_model import LogisticRegression 
# Classification algorithm used in ML suitable for binary classification problem 
from sklearn.metrics import accuracy_score 
#accuracy function is a performance used for evalute the performance for classification by measuring the accuracy of the prediction 
```

```python
df = pd.read_csv(r'C:\Users\sahil\OneDrive\Documents\mail_data.csv')
```

```python
df.head()
```

```python
print(df)
```

```python
data = df.where((pd.notnull(df)), '')
#converting this mail data into normal 
```

```python
data.head()
#printing first few line 
```

```python
data.info()
```

```python
data.loc[data['Category'] =='spam','Category',]=0
data.loc[data['Category'] =='ham','Category',]=1
#if 0 will show as the spam and 1 will show as the ham mail
```

```python
x = data['Message']

y = data['Category']
```

```python
print(x)
```

```python
print(y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state = 3)
#0.2 means it will be 80 and 20% and 80% will be training and 20% will be test/
#and random state is the hyperparameter used to control any randomness invlove in Ml model 
```

```python
print(x.shape)
print(x_train.shape)
print(x_test.shape)
```

```python
print(y.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
feature_extraction = TfidfVectorizer(min_df = 1, stop_words = 'english', lowercase=True)

x_train_features = feature_extraction.fit_transform(x_train)
x_test_features = feature_extraction.transform(x_test)

y_train = y_train.astype('int')
y_test = y_test.astype('int')

#transforming the test data to the feature vector  
#stop words are the english words that dose not add much meaning to the sentence they can easy ignored 
#wihtout sacrificing the meaning of the sentence for eg: words like the, he, has, have 
#
```

```python
print(x_train)
```

```python
print(x_train_features)
```

```python
model = LogisticRegression()
```

```python
#training data 
model.fit(x_train_features, y_train)
```

```python
#the evaluting of the training model and the prediction of the tarin dfata 
prediction_on_training_data = model.predict(x_train_features)
accuracy_on_training_data = accuracy_score(y_train, prediction_on_training_data)
```

```python
print('Acc on training data : ' , accuracy_on_training_data)
```

```python
prediction_on_training_data = model.predict(x_test_features)
accuracy_on_training_data = accuracy_score(y_test, prediction_on_training_data)
```

```python
print('Acc on training data : ' ,  accuracy_on_training_data)
```
